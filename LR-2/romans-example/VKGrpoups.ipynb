```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import vk_requests
from tqdm.auto import tqdm

tqdm.pandas()
sns.set_theme(style="darkgrid")
```

```python
members_hseperm_js = vk_requests.get_users(
    "hseperm",
    fields=["sex", "activities", "bdate", "career", "country", "city", "education", "relation", "universities"],
)
members_hsestudents_js = vk_requests.get_users(
    "hsestudents",
    fields=["sex", "activities", "bdate", "career", "country", "city", "education", "relation", "universities"],
)
```

```python
members_hseperm = pd.DataFrame(members_hseperm_js)
members_hseperm["group"] = "hse_perm"
members_hseperm.head()
```

```python
members_hsestudents = pd.DataFrame(members_hsestudents_js)
members_hsestudents["group"] = "hsestudents"
members_hsestudents.head()
```

```python
members = pd.concat([members_hseperm, members_hsestudents], ignore_index=True)
```

```python
def get_dict_val(df, col, res_col) -> list[str | None]:
    res = []
    for i in df[col]:
        if pd.isna(i):
            res.append(None)
            continue
        res.append(i.get(res_col))
    return res
```

```python
members["city"] = get_dict_val(members, "city", "title")
members["country"] = get_dict_val(members, "country", "title")
```

```python
def transform_bday(row):
    if row is None:
        return None
    if pd.isna(row["bdate"]):
        return None
    if len(row["bdate"].split(".")) < 3:
        return None
    date = pd.to_datetime(row["bdate"], format="%d.%m.%Y")
    return date
```

```python
members["bdate"] = members.apply(transform_bday, axis=1)
```

```python
cat = ["country", "graduation", "education_status", "relation", "sex"]
cat_large = ["city", "university_name"]
```

```python
for c in cat:
    fig, ax = plt.subplots(figsize=(20, 10))
    g = sns.countplot(data=members, x=c, hue="group", ax=ax)
    plt.xticks(rotation=90);
```

```python
for c in cat_large:
    val_count = members[c].value_counts()
    cleared = members.copy()
    cleared[c] = np.where(members[c].isin(val_count.index[val_count > 5]), members[c], "other")
    fig, ax = plt.subplots(figsize=(20, 10))
    g = sns.countplot(data=cleared, x=c, hue="group", ax=ax)
    plt.xticks(rotation=90);
```
